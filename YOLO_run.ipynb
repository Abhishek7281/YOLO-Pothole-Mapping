```python
# Import necessary libraries
import os
import zipfile

# Function to zip a folder
def zip_folder(folder_path, zip_name):
    """
    Compress a folder into a zip file.

    Parameters:
    folder_path (str): Path to the folder to be zipped.
    zip_name (str): Name of the resulting zip file.
    """
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                # Add file to zip, keeping the relative path
                arcname = os.path.relpath(file_path, start=folder_path)
                zipf.write(file_path, arcname)
    print(f"Folder '{folder_path}' has been compressed into '{zip_name}'.")

# Define folder to zip and output zip file name
folder_to_zip = "runs/detect/train11"  # Change this to the folder path
zip_output_name = "yolov11l(132epoch)output.zip"         # Change this to your desired zip file name

# Call the function
zip_folder(folder_to_zip, zip_output_name)

# Verify if the zip file was created
if os.path.exists(zip_output_name):
    print(f"'{zip_output_name}' created successfully!")
else:
    print("Error: Zip file was not created.")
```
